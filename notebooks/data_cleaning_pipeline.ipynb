## Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import glob
import os

pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

## Step 2: Define Columns to Keep
Selecting only required columns at load time to minimize RAM usage.

```python
columns_to_keep = [
    # Core Identifiers & Dates
    'FlightDate', 'Month', 'DayofMonth', 'DayOfWeek',
    'Reporting_Airline', 'Tail_Number', 'Flight_Number_Reporting_Airline',
    'Origin', 'Dest', 'Distance',

    # Departure Metrics
    'CRSDepTime', 'DepDelay', 'TaxiOut', 'WheelsOff',

    # Arrival Metrics
    'CRSArrTime', 'ArrDelay', 'TaxiIn',

    # Duration Metrics
    'CRSElapsedTime', 'ActualElapsedTime', 'AirTime',

    # Filtering Columns (dropped after use)
    'Cancelled', 'Diverted'
]
```

## Step 3: Load and Combine Raw Datasets
Loads all CSVs from the `data/` directory. Uses `usecols` for memory efficiency.

```python
path_to_files = os.path.join(os.getcwd(), '..', 'data', 'flight_data_*.csv')
file_names = glob.glob(path_to_files)

if not file_names:
    raise FileNotFoundError(f"No CSV files found at: {path_to_files}")

df_list = [pd.read_csv(file, usecols=columns_to_keep) for file in file_names]
df = pd.concat(df_list, ignore_index=True)

print(f"\nInitial data loaded. Shape: {df.shape}")
df.head()
```

## Step 4: Data Summary

```python
print("Info of the DataFrame:")
df.info()

print("\nDescription of the DataFrame:")
df.describe()
```

## Step 5: NULL and Duplicate Checks

```python
print("Number of missing values per column:")
print(df.isnull().sum())

print("\nNumber of duplicate rows:")
print(df.duplicated().sum())
```

## Step 6: Data Cleaning

```python
initial_rows = len(df)

# --- CHECK 1: DROP CANCELLED AND DIVERTED FIRST ---
before = len(df)
df = df[(df['Cancelled'] == 0) & (df['Diverted'] == 0)]
df = df.drop(columns=['Cancelled', 'Diverted'])

# --- CHECK 2: DROP ROWS WITH MISSING CRITICAL VALUES ---
before = len(df)
critical_columns = ['DepDelay', 'ArrDelay', 'TaxiOut', 'AirTime', 'ActualElapsedTime', 'CRSElapsedTime']
df = df.dropna(subset=critical_columns)

# For Tail_Number, fill missing with 'Unknown'
df['Tail_Number'] = df['Tail_Number'].fillna('Unknown')

# --- CHECK 3: LOGICAL SANITY CHECKS ---
before = len(df)
df = df[(df['Distance'] > 0) & (df['AirTime'] > 0) & (df['ActualElapsedTime'] > 0)]

# --- CHECK 4: DELAY BOUNDS ---
# Upper cap: 1440 min = 24 hours — delays beyond this are likely data errors or extreme outliers
# Lower cap: -60 min — arriving/departing >60 min early is a scheduling
before = len(df)
df = df[
    (df['DepDelay'] >= -60) & (df['DepDelay'] < 1440) &
    (df['ArrDelay'] >= -60) & (df['ArrDelay'] < 1440)
]

print(f"Final rows: {len(df)}")
print(f"Total rows removed: {initial_rows - len(df)} ({(initial_rows - len(df)) / initial_rows * 100:.2f}%)")
```

## Step 7: Feature Engineering

```python
# --- FEATURE 1: Make_Up_Time ---
df['Make_Up_Time'] = df['DepDelay'] - df['ArrDelay']

# Cap outliers: values beyond ±300 min are likely data errors
before = len(df)
df = df[(df['Make_Up_Time'] > -300) & (df['Make_Up_Time'] < 300)]

# --- FEATURE 2: Carrier_Type ---
# Mainline:  Full-service network carriers with hub-and-spoke operations
# ULCC:      Ultra-low-cost point-to-point carriers
# Regional:  Contract carriers operating under mainline brands

mainline_carriers = ['AA', 'DL', 'UA', 'WN', 'AS', 'B6', 'HA']
ulcc_carriers     = ['NK', 'F9', 'G4']

def classify_carrier(code):
    if code in mainline_carriers:
        return 'Mainline'
    elif code in ulcc_carriers:
        return 'ULCC'
    else:
        return 'Regional'

df['Carrier_Type'] = df['Reporting_Airline'].map(classify_carrier)

# --- FEATURE 3: Time_Block ---
hour = df['CRSDepTime'] // 100
conditions = [
    (hour >= 5)  & (hour < 12),
    (hour >= 12) & (hour < 17),
    (hour >= 17) & (hour < 21),
]
choices = ['Morning', 'Afternoon', 'Evening']
df['Time_Block'] = np.select(conditions, choices, default='Night')
```

## Step 8: Optimize Data Types

```python
# FlightDate to datetime
df['FlightDate'] = pd.to_datetime(df['FlightDate'])

# Convert categorical strings to 'category' dtype
categorical_cols = ['Reporting_Airline', 'Origin', 'Dest', 'Tail_Number', 'Carrier_Type', 'Time_Block']
for col in categorical_cols:
    df[col] = df[col].astype('category')

# Downcast 64-bit floats to 32-bit
float_cols = df.select_dtypes(include=['float64']).columns
df[float_cols] = df[float_cols].astype('float32')

# Downcast 64-bit ints to 32-bit
int_cols = df.select_dtypes(include=['int64']).columns
df[int_cols] = df[int_cols].astype('int32')

print(df.info(memory_usage='deep'))
```

## Step 9: Final Export

```python
output_path = os.path.join(os.getcwd(), '..', 'data', 'cleaned_flight_data_Q1_2024.parquet')
df.to_parquet(output_path, index=False)

print(f"Success! Saved to: {output_path}")
print(f"Final shape: {df.shape}")
```
